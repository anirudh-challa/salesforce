```python
arr_train[0,0:8]
```

```python
train_data = arr_train[:,0:8]
train_label = arr_train[:,8]
test_data = arr_test[:,0:8]
test_label = arr_test[:,8]
```

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(train_data, train_label) 
ans = clf.predict(test_data)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_label,ans)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=2, random_state=0).fit(train_data, train_label)
```
